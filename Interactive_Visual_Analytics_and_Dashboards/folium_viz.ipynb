## Task 2: Mark the success/failed launches for each site on the map

Next, let's try to enhance the map by adding the launch outcomes for each site, and see which sites have high success rates. Recall that data frame spacex_df has detailed launch records, and the `class` column indicates if this launch was successful or not

```python
spacex_df.tail(10)
```

Next, let's create markers for all launch records. If a launch was successful (class=1), then we use a green marker and if a launch was failed, we use a red marker (class=0)

Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.

Let's first create a `MarkerCluster` object

```python
marker_cluster = MarkerCluster()
```

TODO: Create a new column in `launch_sites` dataframe called `marker_color` to store the marker colors based on the `class` value

```python
# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red

def assign_marker_color(class_value):
    if class_value == 1:
        marker_color = 'green'
    else:
        marker_color = 'red'
    return marker_color

launch_sites_df['marker_color'] = spacex_df['class'].apply(assign_marker_color)
```

```python
# Function to assign color to launch outcome
def assign_marker_color(launch_outcome):
    if launch_outcome == 1:
        return 'green'
    else:
        return 'red'
    
spacex_df['marker_color'] = spacex_df['class'].apply(assign_marker_color)
spacex_df.tail(10)
```

## TASK 3: Calculate the distances between a launch site to its proximities

Next, we need to explore and analyze the proximities of launch sites.

Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)

```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

Now zoom in to a launch site and explore its proximity to see if you can easily find any railway, highway, coastline, etc. Move your mouse to these points and mark down their coordinates (shown on the top-left) in order to the distance to the launch site.

You can calculate the distance between two points on the map based on their Lat and Long values using the following method:

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

TODO: Mark down a point on the closest coastline using MousePosition and calculate the distance between the coastline point and the launch site.

```python
# find coordinate of the closet coastline
coastline_lat = 28.564250
coastline_lon = -80.56811

launch_site_lat = 28.562302
launch_site_lon = -80.577356
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
distance_coastline
```

TODO: After obtained its coordinate, create a folium.Marker to show the distance

```python
# Create and add a folium.Marker on your selected closest coastline point on the map
# Display the distance between coastline point and launch site using the icon property 
# for example
# distance_marker = folium.Marker(
#    coordinate,
#    icon=DivIcon(
#        icon_size=(20,20),
#        icon_anchor=(0,0),
#        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance),
#        )
#    )

distance_marker = folium.Marker([coastline_lat, coastline_lon],
                                icon=DivIcon(icon_size=(20,20),
                                             icon_anchor=(0,0),
                                             html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_coastline)))
```

TODO: Draw a PolyLine between a launch site to the selected coastline point

```python
# Create a `folium.PolyLine` object using the coastline coordinates and launch site coordinate
# lines=folium.PolyLine(locations=coordinates, weight=1)
# site_map.add_child(lines)

lines = folium.PolyLine(locations=[[coastline_lat, coastline_lon], [launch_site_lat, launch_site_lon]], weight=1).add_child(folium.Popup('Distance: {:10.2f} KM'.format(distance_coastline)))
site_map.add_child(lines)
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
```

```python
from geopy.distance import geodesic

# Coordenadas del sitio de lanzamiento
launch_site_coords = [launch_sites_df['Lat'][0], launch_sites_df['Long'][0]]

# Coordenadas de la ciudad, autopista y lnea costera ms cercana:
city_lat = 28.40046
city_long = -80.60394
railway_lat = 28.57233
railway_long = -80.58525
highway_lat = 28.4954
highway_long = -80.58249

# Coordenadas de la ciudad, autopista y lnea costera ms cercana
closest_city_coords = [city_lat, city_long]  # Reemplazar con las coordenadas reales
closest_highway_coords = [highway_lat, highway_long]  # Reemplazar con las coordenadas reales
closest_railway_coords = [railway_lat, railway_long]  # Reemplazar con las coordenadas reales

# Calcular las distancias en kilmetros
city_distance = geodesic(launch_site_coords, closest_city_coords).km
highway_distance = geodesic(launch_site_coords, closest_highway_coords).km
coastline_distance = geodesic(launch_site_coords, closest_railway_coords).km

# Crear un marcador para cada ubicacin con la distancia en el popup
city_marker = folium.Marker(
    closest_city_coords, 
    popup=f'Closest City, Distance: {city_distance} km',
    icon=folium.Icon(color='blue', icon='cloud')
)
highway_marker = folium.Marker(
    closest_highway_coords, 
    popup=f'Closest Highway, Distance: {highway_distance} km',
    icon=folium.Icon(color='green', icon='road')
)
coastline_marker = folium.Marker(
    closest_railway_coords, 
    popup=f'Closest Railway, Distance: {coastline_distance} km',
    icon=folium.Icon(color='red', icon='rail')
)

# Dibujar una lnea entre el sitio de lanzamiento y cada ubicacin
line_to_city = folium.PolyLine(locations=[launch_site_coords, closest_city_coords], color='blue')
line_to_highway = folium.PolyLine(locations=[launch_site_coords, closest_highway_coords], color='green')
line_to_railway = folium.PolyLine(locations=[launch_site_coords, closest_railway_coords], color='red')

# Aadir los elementos al mapa
site_map.add_child(city_marker)
site_map.add_child(highway_marker)
site_map.add_child(coastline_marker)
site_map.add_child(line_to_city)
site_map.add_child(line_to_highway)
site_map.add_child(line_to_railway)

# Mostrar el mapa
site_map
```
